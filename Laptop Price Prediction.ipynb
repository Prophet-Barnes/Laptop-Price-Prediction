```python
import pandas as pd, numpy as np
```

```python
df = pd.read_csv("laptop_price.csv")
df
```

#### # laptop_ID is of no use. Removing it:

```python
df.pop('laptop_ID') #df.drop() also works, used ahead in this notebook
```

### Importing Regular Expression (RegEx) for extracting only integers out of some columns of the dataset:

```python
import re
df['Memory'] = [re.sub("[^0-9]", "", df['Memory'][i]) for i in range(0,len(df['Memory']))]
df
```

```python
df.info()
```

```python
df.info()
```

### Strip function can also be used as an alternative to Regular Expression in our case, as input features in some columns are written in a repeating pattern.

```python
df['Ram'] = [df['Ram'][i].rstrip('GB') for i in range(0,len(df['Ram']))]
```

```python
df['Weight'] = [df['Weight'][i].rstrip('kg') for i in range(0,len(df['Weight']))]
df
```

```python
df['ScreenResolution'].describe()
```

```python
ScreenType = []
for i in range(0,len(df['ScreenResolution'])):
    Screen = df['ScreenResolution'][i].split()
    ScreenType.append(' '.join(Screen[0:-1]))
    
ScreenType
```

```python
len(ScreenType)
```

```python
df['ScreenType'] = ScreenType
df['ScreenType'].value_counts()
```

#### The data for this column can still  be simplified. But an understanding about the Screen Types is necessary, so as to determine which detail about the Screen Type should be retained and which should be discarded from the column.

#### For now, we will be discarding this column so as to keep our main focus on our Regression Model. We will only be retaining the Pixel Values for the Screen Resolution.

```python
df.drop(columns = 'ScreenType', inplace = True)
```

```python
Pixels = []
for i in range(0,len(df['ScreenResolution'])):
    Screen = df['ScreenResolution'][i].split()
    Pixels.append(Screen[-1])

Pixels
```

```python
len(Pixels)
```

```python
pixels_l = []
pixels_w = []

for i in range(0,len(Pixels)):
    pixels_l.append(Pixels[i].split('x')[0])
    pixels_w.append(Pixels[i].split('x')[1])
```

```python
pixels_l
```

```python
pixels_w
```

```python
df['pixels_l'] = pixels_l
df['pixels_w'] = pixels_w
df.pop('ScreenResolution')
df
```

```python
cpu = []
for i in range(0,len(df['Cpu'])):
    processor = df['Cpu'][i].split()
    cpu.append(' '.join(processor[0:3]))

cpu
```

```python
df['Cpu'] = cpu
df['Cpu'].value_counts()
```

```python
df['TypeName'].value_counts()
```

```python
df['Product'].value_counts()
```

```python
df['OpSys'].value_counts()
```

```python
df['Gpu'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cat_col = ['Company', 'Product', 'TypeName', 'Cpu', 'Gpu', 'OpSys']
for each_col in cat_col:
    le = LabelEncoder()
    df[each_col] = le.fit_transform(df[each_col])
    
df
```

```python
df.info()
```

```python
for i in ['Ram','Memory','pixels_l','pixels_w']:
    df[i] = df[i].astype(int)
    
df['Weight'] = df['Weight'].astype(float)
df.info()
```

### Visualising the data:

```python
import matplotlib.pyplot as plt, seaborn as sb
```

```python
sb.heatmap(df.corr())
```

```python
x = df[['Company', 'Product', 'TypeName', 'Inches', 'Cpu', 'Ram', 'Memory',
       'Gpu', 'OpSys', 'Weight', 'pixels_l', 'pixels_w']].values
y = df['Price_euros'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train,y_train)    
y_pred = model.predict(x_test)
y_pred   #predicted output 
```

```python
plt.figure(figsize=(15,5))
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
```
